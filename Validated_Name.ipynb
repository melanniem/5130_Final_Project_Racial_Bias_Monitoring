```python
if 'Unnamed: 0' in df.columns:
    df = df.drop('Unnamed: 0', axis=1)

df_clean = df[['first', 'last', 'name', 'identity', 'w.asian', 'mean.correct']].copy()

df_clean.head(3)
```

```python
# White
white_all = df_clean[df_clean['identity'] == 'White'].copy()
print(f"\nWhite: {len(white_all)} names")

# Black
black_all = df_clean[df_clean['identity'] == 'Black or African American'].copy()
print(f"Black or African American: {len(black_all)} names")

# Hispanic
hispanic_all = df_clean[df_clean['identity'] == 'Hispanic'].copy()
print(f"Hispanic: {len(hispanic_all)} names")

# Asian (not White Asian)
asian_all = df_clean[(df_clean['identity'] == 'Asian or Pacific Islander') &
                      (df_clean['w.asian'] == 0)].copy()
print(f"Asian or Pacific Islander: {len(asian_all)} names")

# White Asian
asian_white = df_clean[(df_clean['identity'] == 'Asian or Pacific Islander') &
                        (df_clean['w.asian'] == 1)].copy()
print(f"White Asian: {len(asian_white)} names")
```

```python
threshold = 0.65
white_high = white_all[white_all['mean.correct'] > threshold].sort_values('mean.correct', ascending=False)
black_high = black_all[black_all['mean.correct'] > threshold].sort_values('mean.correct', ascending=False)
hispanic_high = hispanic_all[hispanic_all['mean.correct'] > threshold].sort_values('mean.correct', ascending=False)
asian_high = asian_all[asian_all['mean.correct'] > threshold].sort_values('mean.correct', ascending=False)

print(f"White: {len(white_high)} names (Average rate: {white_high['mean.correct'].mean():.1%})")
print(f"Black: {len(black_high)} names (Average rate: {black_high['mean.correct'].mean():.1%})")
print(f"Hispanic: {len(hispanic_high)} names (Average rate: {hispanic_high['mean.correct'].mean():.1%})")
print(f"Asian: {len(asian_high)} names (Average rate: {asian_high['mean.correct'].mean():.1%})")

print(f"\nTotal: {len(white_high) + len(black_high) + len(hispanic_high) + len(asian_high)} names")
```

```python

print("\n WHITE (Top 10):")
for idx, row in white_high.head(10).iterrows():
    print(f"  {row['name']:25s} - {row['mean.correct']:.1%}")

print("\n BLACK (Top 10):")
for idx, row in black_high.head(10).iterrows():
    print(f"  {row['name']:25s} - {row['mean.correct']:.1%}")

print("\n HISPANIC (Top 10):")
for idx, row in hispanic_high.head(10).iterrows():
    print(f"  {row['name']:25s} - {row['mean.correct']:.1%}")

print("\n ASIAN (Top 10):")
for idx, row in asian_high.head(10).iterrows():
    print(f"  {row['name']:25s} - {row['mean.correct']:.1%}")
```

```python
all_racial_markers = pd.concat([white_high, black_high, hispanic_high, asian_high])
all_racial_markers.to_csv('racial_markers.csv', index=False)
```
